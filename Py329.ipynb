```python
# match STatement

# match variable:
#   case value1:
#     sb
#   case value2:
#     sb
#   case value3:
#     sb
#   .........
#   case _:
#     default sb
```

```python
num1=int(input("Enter the Number 1 : "))
num2=int(input("Enter the Number 2 : "))

oper=input("Enter the Process - +, -, *, /, ** : ")

match oper:
  case "+":
    print("Answer : ",num1+num2)
  case "-":
    print("Answer : ",num1-num2)
  case "*":
    print("Answer : ",num1*num2)
  case "/":
    print("Answer : ",num1/num2)
  case "**":
    print("Answer : ",num1**num2)
  case _:
    print("Please Enter the Value +,-,*,/,**")
```

```python
# Multilevel Inheritance - A child class inherits from another child class


class GrandParent:
  def gpMethod(self):
    print("Grand Parent")

class Parent(GrandParent):
  def pMethod(self):
    print("Parent")

class Child(Parent):
  def cMethod(self):
    print("Child")

C=Child()
C.gpMethod()
C.pMethod()
C.cMethod()
```

```python
# Multiple Inheritance - one child class inherits from multiple parent classes

class Father:
  def fatherProperty(self):
    print("Land")

class Mother:
  def motherProperty(self):
    print("Gold")

class Child(Father, Mother):
  def childProperty(self):
    print("Bike")

C=Child()
C.fatherProperty()
C.motherProperty()
C.childProperty()
```

```python
# Hierarchical Inheritance = multiple child classes inherit from a single parent class

class Animal:
  def eat(self):
    print("Eating")

class Dog(Animal):
  def sound(self):
    print("Dog Barks")

class Cat(Animal):
  def sound(self):
    print("Cats Memow")

D=Dog()
D.eat()
D.sound()

C=Cat()
C.eat()
C.sound()
```

```python
# Hybrid Inheritance - Combination of multiple inheritance types

class A:
  def mehtodA(self):
    print("Class A")

class B(A):
  def methodB(self):
    print("Class B")

class C(A):
  def methodC(self):
    print("Class C")

class D(B,C):
  def methodD(self):
    print("Class D")

obj=D()
obj.mehtodA()
obj.methodB()
obj.methodC()
obj.methodD()
```

```python
# Constructor Inheritance

class Parent:
  def __init__(self):
    print("Parent Constructor")

class Child(Parent):
  def __init__(self):
    Parent.__init__(self)
    print("Child Constructor")
C=Child()
```

```python
# using super() - function is used to call parent class methods and constructor

class Parent:
  def __init__(self):
    print("Parent Constructor")

class Child(Parent):
  def __init__(self):
    super().__init__()
    print("Child Constructor")
C=Child()
```

```python
# Abstraction -

# What is Abstraction - Hiding Implementation Details and Showing Only Essential Functionality to the User

# Python Provides abstraction using the abc (Abstract Base Class) Modules

# Why Abstraction?

# Hides Complex Implementation, Improves Security, Reduces Code Complexity , Enforces Rules in Child Classes
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
  @abstractmethod
  def start(self):
    pass

class Car(Vehicle):
  def start(self):
    print("Car Started")

C=Car()
C.start()
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def area(self):
    print("Area of Circle")

obj=Circle()
obj.area()
```

```python
from abc import ABC, abstractmethod

class Employee(ABC):
  @abstractmethod
  def salary(self):
    pass

class Manager(Employee):
  def salary(self):
    print("Salary = 50000")

class Developer(Employee):
  def salary(self):
    print("Salary = 40000")

M=Manager()
D=Developer()

M.salary()
D.salary()
```

```python
from abc import ABC, abstractmethod

class Payment(ABC):
  @abstractmethod
  def pay(self, amount):
    pass

class GooglePay(Payment):
  def pay(self, amount):
    print("Paid", amount, "using Google Pay")

class PhonePay(Payment):
  def pay(self, amount):
    print("Paid", amount, "Using PhonePay")

G=GooglePay()
P=PhonePay()


G.pay(4000)
P.pay(1000)
```

```python
# Overriding Methods

# What is Method Overriding?

# Method Overriding occurs when a child class provides its own implementation of a method already defined in the parent class


class Animal:
  def sound(self):
    print("Animal Sound")

class Dog(Animal):
  def sound(self):
    print("Bark")
D=Dog()
D.sound()
```

```python
class Person:
  def display(self):
    print("Person Details")

class Student(Person):
  def display(self):
    super().display()
    print("Student Details")

S=Student()
S.display()
```

```python
class Vehicle:
  def start(self):
    print("Vehicle Started")

class Car(Vehicle):
  def start(self):
    super().start()
    print("Car Started")

C=Car()
C.start()
```

```python
class Number:
  def __init__(self, value):
    self.value=value

  def __add__(self, other):
    return self.value+other.value

n1=Number(10)
n2=Number(20)

print(n1+n2)
```

```python
# Public Variables

class Student:
  def __init__(self):
    self.name="Sridhar"

S=Student()
print(S.name)
```

```python
# Protected Variables

class Student:
  def __init__(self):
    self._name="Kumar"

S=Student()
print(S._name)
```

```python
# Private Variables

class Student:
  def __init__(self):
    self.__name="Sridhar Raj P"
  def display(self):
    print(self.__name)

S=Student()
S.display()
```

```python
class Bank:
  def __init__(self):
    self.__balance=100000
  def showBalance(self):
    print("Balance = ", self.__balance)
B=Bank()
B.showBalance()
```

```python
# Getter - Used to Retrieve Data
# Setter - Used to Update Data

class Employee:
  def __init__(self):
    self.__salary=0
  def setSalary(self, salary):
    self.__salary=salary
  def getSalary(self):
    return self.__salary

E=Employee()
E.setSalary(60000)
print(E.getSalary())
```

```python
class ATM:
  def __init__(self):
    self.__balance=10000
  def deposit(self, amount):
    self.__balance+=amount
  def withdraw(self, amount):
    if(amount<=self.__balance):
      self.__balance-=amount
    else:
      print("Insufficient Balance")

  def showBalance(self):
    print("Balance = ", self.__balance)

A=ATM()

A.withdraw(20000)
A.showBalance()
```

```python
class Calculator:
  def add(self, a=0,b=0,c=0):
    print("Result = ", a+b+c)

answer=Calculator()
answer.add(10)
answer.add(10,20)
answer.add(10,20,30)
```

```python
class Calculator:
  def add(self, *numbers):
    print("Sum = ", sum(numbers))

C=Calculator()

C.add(100)
C.add(600,100,200)
C.add(10,20,40,100,200,3000)
```

```python
# Duck Typing

class Student:
  def display(self):
    print("Student Details")

class Employee:
  def display(self):
    print("Employee Details")

def show(obj):
  obj.display()

show(Student())
show(Employee())
```
