```python
#Grade Calculator (if-elif-else)

mark = int(input("Enter mark: "))

if mark >= 90:
    print("Grade A")
elif mark >= 75:
    print("Grade B")
elif mark >= 50:
    print("Grade C")
else:
    print("Fail")
```
